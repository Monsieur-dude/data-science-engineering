```python
def print_5xn(string) :
    chunk_num = int(len(string)/5)
    
    for i in range(chunk_num+1) :
        print(string[i*5:i*5 + 5])
        
print_5xn("아이엠어보이유알어걸스")
```

```python
def print_mxn(string,number) :
    chunk_num = int(len(string)/number)
    for i in range(chunk_num + 1) :
        print(string[i*number : i*number + number])

print_mxn("아이엠어보이유알어걸", 6)
```

```python
def make_list(string) :
    list_empty = []
    for i in range(len(string)) :
        list_empty.append(i)
    print(list_empty)
make_list("string")
```

```python
def pickup_even(list) :
    list_even = []
    for i in list :
        if i % 2 == 0 :
            list_even.append(i)
    return list_even
pickup_even([3,5,4,2,6,5,8,9,0,6])
```
