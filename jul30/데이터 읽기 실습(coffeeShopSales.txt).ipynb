```python
## 데이터 헤더 처리

f = open(file_name)
header = f.readline()

## 항목 이름을 리스트로 만들기 .split()

header_list = header.split()

## 헤더를 제외한 데이터를 리스트로 만들기

espresso = []
americano = []
cappucino = []
cafelatte = []

for line in f :
    data_list = line.split()

    espresso.append(int(data_list[1]))      # 일별 판매량를 각 항목으로 갖는 리스트
    americano.append(int(data_list[2]))     # 일별 판매량를 각 항목으로 갖는 리스트
    cappucino.append(int(data_list[3]))     # 일별 판매량를 각 항목으로 갖는 리스트
    cafelatte.append(int(data_list[4]))     # 일별 판매량를 각 항목으로 갖는 리스트

## for를 이용하여 총 판매량, 하루 평균 판매량 구하기

total_sum = [sum(espresso), sum(americano), sum(cappucino), sum(cafelatte)]
total_mean = [sum(espresso)/len(espresso), sum(americano)/len(americano), sum(cappucino)/len(cappucino), sum(cafelatte)/len(cafelatte)]

for k in range(len(total_sum)) :
    print("{0} 판매량".format(header_list[k+1]))
    print(" - 4일 전체 : {0}잔, 하루 평균 : {1}잔".format(total_sum[k], total_mean[k]))
print("---------------------------------------------","\n", "4일간 총 커피 판매량 {}잔, 하루 평균 판매량 {}잔".format(sum(total_sum), sum(total_mean)))

f.close()
```

```python
with open("coffeeShopSales.txt", "r") as f :
    header = f.readline()
    header_list = header.split()

    espresso = []
    americano = []
    cappucino = []
    cafelatte = []

    for line in f :
        data_list = line.split()

        espresso.append(int(data_list[1]))      
        americano.append(int(data_list[2]))     
        cappucino.append(int(data_list[3]))     
        cafelatte.append(int(data_list[4]))     

    total_sum = [sum(espresso), sum(americano), sum(cappucino), sum(cafelatte)]
    total_mean = [sum(espresso)/len(espresso), sum(americano)/len(americano), sum(cappucino)/len(cappucino), sum(cafelatte)/len(cafelatte)]

    for k in range(len(total_sum)) :
        print("{0} 판매량".format(header_list[k+1]))
        print(" - 4일 전체 : {0}잔, 하루 평균 : {1}잔".format(total_sum[k], total_mean[k]))
    
    print("---------------------------------------------","\n", "4일간 총 커피 판매량 {}잔, 하루 평균 판매량 {}잔".format(sum(total_sum), sum(total_mean)))
```
